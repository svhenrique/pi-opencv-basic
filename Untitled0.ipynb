```python
# convertendo BGR para LAB
image = cv2.cvtColor(original_image, cv2.COLOR_BGR2LAB)
cv2_imshow(image)
```

```python
# convertendo BGR para CMYK

# Create float
image = original_image.astype(float)/255.

# Extract channels
with np.errstate(invalid='ignore', divide='ignore'):
	K = 1 - np.max(image, axis=2)
	C = (1-image[...,2] - K)/(1-K)
	M = (1-image[...,1] - K)/(1-K)
	Y = (1-image[...,0] - K)/(1-K)

# Convert the input BGR image to CMYK colorspace
CMYK = (np.dstack((C,M,Y,K)) * 255).astype(np.uint8)

# Saving image
image = Image.fromarray(CMYK, 'CMYK')
image.save(image_path + 'cmyk.tiff')
```

```python
# lendo nova imagem
woman = cv2.imread(image_path + 'woman.jpg')
woman = cv2.resize(woman, None, fx=.52, fy=.52)
```

```python
# mostrando imagem para segmentação
cv2_imshow(woman)
```

```python
# convertendo BGR para RGB (pois trabalharemos com gráficos do plt)
woman = cv2.cvtColor(woman, cv2.COLOR_BGR2RGB)
```

```python
# convertendo para HSV
hsv_woman = cv2.cvtColor(woman, cv2.COLOR_BGR2HSV)
```

```python
# verificando valores em RGB

# separando canais de cores
r, g, b = cv2.split(woman)
ax = plt.axes(projection='3d')

# normalizando valores de cores
pixel_colors = woman.reshape((woman.shape[0]*woman.shape[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()

# criando gráfico
ax.scatter3D(r, g, b, facecolors=pixel_colors, marker=".")
ax.set_xlabel("R")
ax.set_ylabel("G")
ax.set_zlabel("B")
plt.legend()
plt.show()
```

```python
# verificando valores em HSV

# separando canais de cores
h, s, v = cv2.split(hsv_woman)
ax = plt.axes(projection='3d')

# criando gráfico
ax.scatter3D(h, s, v, facecolors=pixel_colors, marker=".")
ax.set_xlabel("H")
ax.set_ylabel("S")
ax.set_zlabel("V")
plt.legend()
plt.show()
```

```python
# preparando máscara amarela
c1 = np.array([50, 100, 200])
c2 = np.array([150, 250, 255])
mask = cv2.inRange(hsv_woman, c1, c2)
```

```python
# passando resultado 
res = cv2.bitwise_and(woman, woman, mask=mask)
```

```python
# mostrando resultado
plt.subplot(1, 3, 1)
plt.title('ORIGINAL')
plt.imshow(woman)
plt.subplot(1, 3, 2)
plt.title('MASK')
plt.imshow(mask, cmap="gray")
plt.subplot(1, 3, 3)
plt.imshow(res)
plt.title('RESULT')
plt.show()
```

### Simples

```python
# carregando imagem para Limiarização Simples
gradient = cv2.imread(image_path + 'gradient.png')
gradient = cv2.resize(gradient, None, fx=.48, fy=.48)
gradient = cv2.cvtColor(gradient, cv2.COLOR_BGR2GRAY)
cv2_imshow(gradient)
```

```python
# separando resultados de limiarização
ret,thresh1 = cv2.threshold(gradient,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(gradient,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(gradient,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(gradient,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(gradient,127,255,cv2.THRESH_TOZERO_INV)
```

#### Binário

*    Se o valor do pixel for maior que o limite atribuído, o pixel receberá o valor do limite, se não, receberá 0 (preto).

```python
cv2_imshow(thresh1)
```

#### Binário inverso

*    Se o valor do pixel for maior que o limite atribuído, o pixel receberá 0, se não, receberá 255.

```python
cv2_imshow(thresh2)
```

#### Truncado 

*    Se o valor do pixel for maior que o limite atribuído, ele receberá o valor do limite. 

```python
cv2_imshow(thresh3)
```

#### Para 0

*    Se o valor do pixel for menor que o limite atribuído, receberá o valor 0. 

```python
cv2_imshow(thresh4)
```

#### Para 0 invertido

*    Se o valor do pixel for maior que o limite atribuído, receberá o valor 0. 

```python
cv2_imshow(thresh5)
```

### Adaptável

*    Determina o limite de um pixel baseado em uma região ao redor dele
*    Obtemos limites diferentes para regiões diferentes da mesma imagem
*    Melhores resultados para imagens com iluminação variável

```python
# carregando imagem para Limiarização Adaptável
dices = cv2.imread(image_path + 'bw.png')
dices = cv2.cvtColor(dices, cv2.COLOR_BGR2GRAY)
cv2_imshow(dices)
```

```python
# mostrando limiarização binária simples
ret, thresh = cv2.threshold(dices, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

#### Média C

*    O valor limite é a média da área da vizinhança menos a constante C.

```python
# preparando Limiarização Adaptável Média C
thresh1 = cv2.adaptiveThreshold(dices, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                          cv2.THRESH_BINARY, 11, 5)
cv2_imshow(thresh1)
```

#### Gaussiana C

*    O valor limite é a soma ponderada gaussiana dos valores de vizinhança menos a constante C.

```python
# preparando Limiarização Adaptável Gaussiana C

thresh2 = cv2.adaptiveThreshold(dices, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 11, 5)
cv2_imshow(thresh2)
```

## Distância Euclidiana



*   É a distância entre dois pontos
*   A distância euclidiana entre os pontos P = (p1, p2, …, pn) e Q = (q1, q2, … , qn) num espaço euclidiano n-dimensional, é:


euclidiano.png

```python
# Lendo imagem original e formatando para ficar do mesmo tamanho que a imagem em CMYK
image_path = "/content/drive/My Drive/images/"
original_image = cv2.imread(image_path + 'Lenna.png')
original_image = cv2.resize(original_image, None, fx=0.52, fy=0.52)

# lendo imagem em CMYK e convertendo para array
cmyk_image = Image.open(image_path + 'cmyk.tiff')
cmyk_image = cmyk_image.convert('RGB')
cmyk_image = np.array(cmyk_image)
cmyk_image = cv2.cvtColor(cmyk_image, cv2.COLOR_BGR2RGB)
```

```python
# mostrando informações de imagem original
print('shape: ', original_image.shape)
cv2_imshow(original_image)
```

```python
# mostrando informações de imagem convertida em CMYK e depois convertida para RGB e então BGR
print('shape: ', cmyk_image.shape)
cv2_imshow(cmyk_image)
```

```python
def euclidean(v1, v2):

  if len(v1) != len(v2):
    raise Exception('v1 e v2 precisam ter dimensões iguais')

  result = 0
  for i in range(len(v1)):
    result += (v1[i] - v2[i])**2
  
  result = result**(1/2)
  return result

def comparation(d1, d2):

  if d1.shape != d2.shape:
    raise Exception('d1 e d2 precisam ter dimensões iguais')

  result = []
  shape = d1.shape

  for i in range(shape[0]):
    for k in range(shape[1]):

      d1_values = d1[i][k]
      d2_values = d2[i][k]
      distance = euclidean(d1_values, d2_values)
      result.append(distance)
  
  return result
```

```python
#cmyk_image, original_image = cv2.cvtColor(cmyk_image, cv2.COLOR_BGR2HSV), cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV)
result = comparation(original_image, cmyk_image)

sns.distplot(result, hist=False, rug=False)
plt.xlabel('Porcentagem')
plt.ylabel('Probabilidades')

```
